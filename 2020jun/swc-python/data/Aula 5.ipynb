## Episódio 6 (continuação)

```python
import glob
import numpy
import matplotlib.pyplot

nome_arqs = sorted(glob.glob('inflammation*.csv'))
nome_arqs = nome_arqs[:3]

for nome_arq in nome_arqs:
    print(nome_arq)
    
    dados = numpy.loadtxt(nome_arq, delimiter=',')
    fig = matplotlib.pyplot.figure(figsize=(10, 3))
    eixos1 = fig.add_subplot(1, 3, 1)
    eixos2 = fig.add_subplot(1, 3, 2)
    eixos3 = fig.add_subplot(1, 3, 3)
    
    eixos1.set_ylabel('médias')
    eixos1.plot(numpy.mean(dados, axis=0))

    eixos2.set_ylabel('máximos')
    eixos2.plot(numpy.max(dados, axis=0))

    eixos3.set_ylabel('mínimos')
    eixos3.plot(numpy.min(dados, axis=0))
    
    fig.tight_layout()
    matplotlib.pyplot.show()
```

#### Gerar Estatísticas Compostas

```python
numpy.zeros((60,40))
```

```python
import glob
import numpy
import matplotlib.pyplot

nome_arqs = sorted(glob.glob('inflammation*.csv'))

dados_compostos = numpy.zeros((60,40))

for nome_arq in nome_arqs:
    dados = numpy.loadtxt(nome_arq, delimiter=',')
    dados_compostos = dados_compostos + dados

dados_compostos = dados_compostos / len(nome_arqs)    
    
fig = matplotlib.pyplot.figure(figsize=(10, 3))
eixos1 = fig.add_subplot(1, 3, 1)
eixos2 = fig.add_subplot(1, 3, 2)
eixos3 = fig.add_subplot(1, 3, 3)

eixos1.set_ylabel('médias')
eixos1.plot(numpy.mean(dados_compostos, axis=0))

eixos2.set_ylabel('máximos')
eixos2.plot(numpy.max(dados_compostos, axis=0))

eixos3.set_ylabel('mínimos')
eixos3.plot(numpy.min(dados_compostos, axis=0))

fig.tight_layout()
matplotlib.pyplot.show()
```

### Checking our Data

```python
import glob
import numpy

nome_arqs = sorted(glob.glob('inflammation*.csv'))

dados = numpy.loadtxt(nome_arqs[2], delimiter=',')
```

```python
inflamação_máxima_dia_0 = numpy.max(dados, axis=0)[0]
inflamação_máxima_dia_20 = numpy.max(dados, axis=0)[20]

ok = True
if inflamação_máxima_dia_0 == 0 and inflamação_máxima_dia_20 == 20:
    print('Valores máximos suspeitos!')
    ok = False
if numpy.sum(numpy.min(dados, axis=0)) == 0:
    print('Mínimos zero!')
    ok = False
if ok:
    print('Tudo OK!')
else:
    print('Dados com problemas reportados acima.')
```
